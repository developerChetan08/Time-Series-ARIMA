```python
import pandas as pd
import numpy as np
```

```python
start = pd.to_datetime("2023-01-01")
end = pd.to_datetime("2023-12-31")
range = pd.date_range(start=start, end=end, freq="D")
```

```python
sequential_data = np.random.rand(len(range),1)
```

```python
sequential_data.ravel()
```

```python
df = pd.DataFrame({"Date":range, "Values":sequential_data.ravel()})
```

```python
df.head()
```

```python
df.set_index("Date", inplace=True)
```

```python
df.plot(figsize=(13,5))
```

```python
import statsmodels
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(df["Values"])
result.plot();
```
