```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
from statsmodels.datasets import get_rdataset
```

```python
aip = get_rdataset("AirPassengers").data["value"]
```

```python
plt.plot(aip)
plt.show()
```

```python
best_order = None
best_aic = np.inf
```

```python
for p in range(10):
    for d in range(5):
        for q in range(10):
            try:
                model=ARIMA(aip, order=(p,d,q))
                result=model.fit()
                if result.aic < best_aic:
                    best_aic = result.aic
                    best_order = (p,d,q)
            except:
                continue
```

```python
model = ARIMA(aip, order=(9,2,8))
result = model.fit()
```

```python
forecast_steps = 24
forecast = result.get_forecast(steps=forecast_steps)
```
